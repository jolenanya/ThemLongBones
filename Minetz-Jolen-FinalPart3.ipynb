## Data Set Summary:
#### The dataset is titled the “Goldman Osteometric Data Set” named after the organization (Joanna Jackson Goldman Memorial Prize) that funded a research project that allowed Dr. Auerbach to collect osteometric data from human skeletal collections around the world. This is the link to the dataset:  https://web.utk.edu/~auerbach/GOLD.htm . The first few variables provide information about each case (the facility the skeleton was located and the ID of that skeleton at that location, estimated sex & age of the skeleton, notes, and the purported origin of the skeleton). The data is in the form of skeletal measurements (in millimeters) taken mainly from the long bones (humerus, radius, ulna, femur, tibia, fibula) but several measurements of the pelvis are also included. In addition to the raw measurement variables, there are several derived variables (calculated from the measurements). The supplemental document “Description of the data set and measurements” (GoldMeasures.pdf) that can be downloaded from the website provides detailed descriptions of the variables and the measurements.

```python
###Adapted from IS590 Code for HW6 Solutions 
#For HEATMAP
#For HEATMAP
x = bones['Age'] #Estimated/known age of each case
y = bones['Inst'] #Institution where skeletal sample is housed... not representative of original geo-location
z = bones['Crural'] #Crural index = LTML/LFML = lower leg/upper leg
#x.shape, y.shape, z.shape

# find unique things
import numpy as np
xnames = x.unique()
ynames = y.unique()
znames = np.zeros([len(xnames),len(ynames)])

# loop and fill
for i,xn in enumerate(xnames):
    for j, yn in enumerate(ynames):
        mask = (x == xn) & (y == yn)
        znames[i,j] = z[mask].sum()
#         znames[i,j] = z[mask]
```

# Dashboard Explanation:
#### The heatmap on the left shows the relationship between the age at death of the skeleton and the institution that the skeletal remains are housed. The age category is difficult to visualize since the raw data is neither in consistent nor compatible formats (i.e. some true ages are recorded: 26, while other ages are estimated and provide an age range: 40-50). 
#### The Institution represents where the skeleton was located when the measurements were collection, but does not represent the geographical origin of that skeleton. 
#### The heat scale to the right of the heat map represents the ‘crural index’, which is the ratio of maximum tibia length to maximum femur length (i.e. the lower leg length divided by the upper leg length). 
#### When a cell in the heatmap is clicked, the scatter plot on the right changes accordingly. 
#### The scatter plot shows the relationship between the maximum length of the humerus relative to the maximum length of the femur. If the is no data for the humerus, the femur length exists with a y-value of 0. If multiple humeri have the same femur measurement, their measurements are summed.
